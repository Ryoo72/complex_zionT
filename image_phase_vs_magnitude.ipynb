```python
# helper to load data from PNG image files
import scipy.misc
```

```python
import numpy
# library for plotting arrays
import matplotlib.pyplot
# ensure the plots are inside this notebook, not an external window
%matplotlib inline
```

```python
img_array = scipy.misc.imread("test_image/test.jpg", flatten=True)
img_array.dtype, img_array.shape
```

```python
matplotlib.pyplot.imshow(img_array, cmap="gray", interpolation='None')
```

```python
fft_array = numpy.fft.rfft(img_array)
fft_array.dtype, fft_array.shape
```

```python
# do stuff in frequency domain
f_mag = numpy.abs(fft_array)
f_phase = numpy.exp(1j*numpy.angle(fft_array))
```

```python
i_mag = numpy.fft.irfft(f_mag)
i_mag.dtype, i_mag.shape
```

```python
i_phase = numpy.fft.irfft(f_phase)
i_phase.dtype, i_phase.shape
```

```python
matplotlib.pyplot.imshow(i_mag, cmap="gray", interpolation='None')
```

```python
matplotlib.pyplot.imshow(i_phase, cmap="gray", interpolation='None')
```
